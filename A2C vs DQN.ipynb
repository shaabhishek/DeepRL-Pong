```python
import pandas
import matplotlib.pyplot as plt
import numpy as np
```

```python
reward_df_dqn = pandas.read_csv('final_rewards_dqn')
reward_df_a2c = pandas.read_csv('final_rewards_a2c')
```

```python
reward_df_dqn.info()
```

```python
plt.figure(figsize=(20,12))

# A2C
rollingmean_eplength = reward_df_a2c.EpisodeLength.rolling(100).mean()
plt.plot(np.linspace(0,8400000, len(rollingmean_eplength)), rollingmean_eplength)
# DQN
rollingmean_eplength = reward_df_dqn.EpisodeLength.rolling(50).mean()
plt.plot(np.linspace(0,3000000, len(rollingmean_eplength)), rollingmean_eplength)

fsize=20
plt.xlabel('Training Steps', fontsize=fsize)
plt.ylabel('Episode Length', fontsize=fsize)
plt.legend(['A2C', 'DQN'], fontsize=fsize)
plt.title('Smoothed Episode Lengths', fontsize=fsize)
plt.savefig('Poster/episode_lengths', transparent=False, bbox_inches='tight')
```

```python
plt.figure(figsize=(20,12))

# A2C
rollingmean_rewards = reward_df_a2c.Reward.rolling(100).mean()
plt.plot(np.linspace(0,8400000, len(rollingmean_rewards)), rollingmean_rewards)
# DQN
rollingmean_rewards = reward_df_dqn.Reward.rolling(50).mean()
plt.plot(np.linspace(0,3000000, len(rollingmean_rewards)), rollingmean_rewards)


plt.xlabel('Training Steps', fontsize=fsize)
plt.ylabel('Reward', fontsize=fsize)
plt.legend(['A2C', 'DQN'], fontsize=fsize)
plt.title('Smoothed Rewards', fontsize=fsize)
plt.savefig('Poster/rewards', tranparent=False, bbox_inches='tight')
```
